```python
import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk

# Career Prediction Logic
def calculate_scores(*args):
    try:
        scores = []
        error_messages = []

        for entry_var, subject in zip(entry_vars, subjects):
            input_value = entry_var.get().strip()

            if input_value == "":  # Skip empty fields, don't calculate till they're all filled
                total_score_var.set("")
                average_score_var.set("")
                return

            try:
                value = int(input_value)
                if not (0 <= value <= 100):
                    error_messages.append(f"{subject} score must be between 0 and 100.")
                else:
                    scores.append(value)
            except ValueError:
                error_messages.append(f"{subject} must be a number between 0 and 100.")

        if error_messages:
            return  # Don't update scores if there's an invalid input

        # Calculate total and average score
        total_score = sum(scores)
        average_score = total_score / len(scores) if scores else 0

        total_score_var.set(f"{total_score}")  # Update total score entry
        average_score_var.set(f"{average_score:.2f}")  # Update average score entry

    except Exception as e:
        messagebox.showerror("Error", f"Calculation failed: {e}")

# Input validation (on button click)
def predict_career():
    try:
        scores = []
        error_messages = []

        for entry_var, subject in zip(entry_vars, subjects):
            input_value = entry_var.get().strip()

            if input_value == "":  # Check for any empty fields
                error_messages.append(f"{subject} field cannot be empty.")
                continue

            try:
                value = int(input_value)
                if not (0 <= value <= 100):
                    error_messages.append(f"{subject} score must be between 0 and 100.")
                else:
                    scores.append(value)
            except ValueError:
                error_messages.append(f"{subject} must be a number between 0 and 100.")

        if error_messages:  # Show validation errors if any
            messagebox.showerror("Invalid Input", "\n".join(error_messages))
            return

        total_score = sum(scores)
        average_score = total_score / len(scores)

        if average_score > 85:
            career = "Doctor"
        elif average_score > 70:
            career = "Engineer"
        elif average_score > 55:
            career = "Artist"
        else:
            career = "Other"

        result_text.delete('1.0', 'end')
        result_text.insert('1.0', f"Total Score: {total_score}\n"
                                  f"Average Score: {average_score:.2f}\n"
                                  f"Recommended Career: {career}")
    except Exception as e:
        messagebox.showerror("Error", f"Prediction failed: {e}")

# Reset all input fields
def reset():
    for entry_var in entry_vars:
        entry_var.set("")
    total_score_var.set("")  # Clear total score entry
    average_score_var.set("")  # Clear average score entry
    result_text.delete('1.0', 'end')

# Create the main window
root = tk.Tk()
root.title("Career Aspiration Prediction System")
root.state('zoomed')

# Load and resize background image
bg_image_path = r"C:\Users\niraj\OneDrive\Desktop\Goals-And-Aspirations-Are-Cruical-to-Pre-MBA-Strategy-1024x724.png"
try:
    bg_image = Image.open(bg_image_path)
    bg_image = bg_image.resize((root.winfo_screenwidth(), root.winfo_screenheight()), Image.LANCZOS)
    bg_photo = ImageTk.PhotoImage(bg_image)
    bg_label = tk.Label(root, image=bg_photo)
    bg_label.place(x=0, y=0, relwidth=1, relheight=1)
except FileNotFoundError:
    messagebox.showerror("Error", "Background image not found.")
    root.destroy()

# Frame for input fields (centered in the middle)
frame = tk.Frame(root, bg="white", bd=2, relief="solid", padx=20, pady=10)
frame.place(relx=0.5, rely=0.5, anchor="center")  # Centered in the middle of the screen

# Title label
label1 = tk.Label(frame, text="Student Score Entry", font=("Times New Roman", 20, "bold"), bg="light yellow")
label1.grid(row=0, columnspan=2, pady=(10, 20))

# Subjects and entries (compact layout)
subjects = ["Math", "Physics", "Chemistry", "Biology", "English", "History", "Geography"]
entry_vars = []

for i, subject in enumerate(subjects):
    label = tk.Label(frame, text=f"{subject} (0-100):", font=("Times New Roman", 14), bg="light yellow")
    label.grid(row=i + 1, column=0, padx=5, pady=3, sticky="w")

    entry_var = tk.StringVar()
    entry_var.trace_add("write", calculate_scores)  # Bind trace to calculate scores dynamically
    entry_vars.append(entry_var)

    entry = tk.Entry(frame, font=("Times New Roman", 14), width=10, textvariable=entry_var)
    entry.grid(row=i + 1, column=1, padx=5, pady=3)

# Add Total Score and Average Score fields below Geography
total_score_var = tk.StringVar()
average_score_var = tk.StringVar()

# Total Score Label and Entry
total_score_label = tk.Label(frame, text="Total Score:", font=("Times New Roman", 14), bg="light yellow")
total_score_label.grid(row=len(subjects) + 1, column=0, padx=5, pady=3, sticky="w")
total_score_entry = tk.Entry(frame, font=("Times New Roman", 14), textvariable=total_score_var, state="readonly", width=10)
total_score_entry.grid(row=len(subjects) + 1, column=1, padx=5, pady=3)

# Average Score Label and Entry
average_score_label = tk.Label(frame, text="Average Score:", font=("Times New Roman", 14), bg="light yellow")
average_score_label.grid(row=len(subjects) + 2, column=0, padx=5, pady=3, sticky="w")
average_score_entry = tk.Entry(frame, font=("Times New Roman", 14), textvariable=average_score_var, state="readonly", width=10)
average_score_entry.grid(row=len(subjects) + 2, column=1, padx=5, pady=3)


# Prediction and Reset buttons
button1 = tk.Button(frame, text="PREDICT", command=predict_career, font=("Times New Roman", 16, "bold"),
                    bg="black", fg="coral")
button1.grid(row=len(subjects) + 3, column=0, pady=(10, 0))

button2 = tk.Button(frame, text="CLEAR", command=reset, font=("Times New Roman", 16, "bold"),
                    bg="black", fg="coral")
button2.grid(row=len(subjects) + 3, column=1, pady=(10, 0))

# Text widget for results (positioned at the bottom center)
result_text = tk.Text(root, height=4, width=50, font=("Times New Roman", 16), bg="lemon chiffon", fg="dark blue")
result_text.place(relx=0.5, rely=0.8, anchor="center")

# Start the Tkinter main loop
root.mainloop()
```
